<h1>Dendograma Hieraquicos</h1>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.preprocessing import StandardScaler
```

<h3>1. Entrando com os dados</h3>

```python
dados = pd.read_csv('storage/processed/dados_mean.csv', sep=',')

dados
```

<h3>2. Normalização dos Dados</h3>

```python
# Normalizando os dados (se necessário)
scaler = StandardScaler()
dados_padronizados = scaler.fit_transform(dados)

dados_padronizados = pd.DataFrame(dados_padronizados)

dados_padronizados
```

<h3>3. Aplicação do Algoritimo de Agrupamento Hierarquico</h3>

```python
# Aplicando o algoritmo de aglomeração hierárquica
resultado = linkage(dados_padronizados, method='ward')  # 'ward' é um método de aglomeração hierárquica

resultado
```

<h3>4. Visualização do Dendrograma</h3>

```python
# Gerando o dendrograma
plt.figure(figsize=(12, 7))

dendrogram(resultado)
plt.title('Dendrograma de Aglomeração Hierárquica')
plt.xlabel('Índices das Amostras')
plt.ylabel('Distância')
plt.show()
```

```python
plt.figure(figsize=(12, 7))
dendrogram(resultado)
plt.axhline(y=50, color='r', linestyle='--')  # Linha de corte em 20
plt.title('Dendrograma com Linha de Corte')
plt.xlabel('Índices das Amostras')
plt.ylabel('Distância')
plt.show()
```

<h3>Construção de Cluster</h3>

```python
from scipy.cluster.hierarchy import fcluster

# Definindo o número de clusters a partir da altura de corte (por exemplo, corte em 20)
clusters = fcluster(resultado, t=40, criterion='distance')

# Adicionando os clusters ao DataFrame
dados['cluster'] = clusters

# Exibindo os resultados
dados
```
