<h1>Aplicação do KNN para construção de Scores para Clientes</h1>

```python
from sklearn.neighbors import NearestNeighbors
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd 
import warnings

warnings.filterwarnings('ignore') 
```

```python
data = pd.read_csv('storage/processed/dados_mean.csv', sep=',')

data
```

```python
# Padronizando as variáveis
scaler = StandardScaler()
dados_padronizados = scaler.fit_transform(data)

dados_padronizados = pd.DataFrame(dados_padronizados)

dados_padronizados
```

```python
# Usando KNN para encontrar os vizinhos mais próximos
knn = NearestNeighbors(n_neighbors=2)
knn.fit(dados_padronizados)

# Calculando as distâncias para os vizinhos mais próximos
distances, _ = knn.kneighbors(dados_padronizados)
```

```python
plt.plot(distances)
```

```python
# Criando um score baseado na soma das distâncias
score = np.mean(distances, axis=1)

# Normalizando o score para a faixa de 0 a 100
score_normalized = 100 * (score - score.min()) / (score.max() - score.min())
print(score_normalized)
```

```python
plt.plot(score_normalized)
```

```python
data['score'] = score_normalized

data
```
